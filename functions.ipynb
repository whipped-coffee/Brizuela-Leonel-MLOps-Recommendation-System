### def peliculas_idioma( Idioma: str ) ###
 Se ingresa un idioma (como están escritos en el dataset, no hay que traducirlos!). Debe devolver la cantidad de películas producidas en ese idioma.
                    Ejemplo de retorno: X cantidad de películas fueron estrenadas en idioma

```python
def movies_language(language):
    movie_counts = {f'count movies produced in {language}': data.loc[data['release_language'] == str(language), 'id'].count()}
    return movie_counts
movies_language('English')
```

### def peliculas_duracion( Pelicula: str ): ### 
Se ingresa una pelicula. Debe devolver la la duracion y el año.
                    Ejemplo de retorno: X . Duración: x. Año: xx

```python
def movies_runtime(movie):
    movie_data = {'movie': movie,
                  'runtime': data.loc[data['title'] == str(movie), 'runtime'].values[0].astype(int),
                  'release_year': data.loc[data['title'] == str(movie), 'release_year'].values[0]}
    return movie_data
movies_runtime('Toy Story')
```

### def franquicia( Franquicia: str ) ###
: Se ingresa la franquicia, retornando la cantidad de peliculas, ganancia total y promedio
                    Ejemplo de retorno: La franquicia X posee X peliculas, una ganancia total de x y una ganancia promedio de xx

```python
def collection(movie):
    if (data.loc[data['collection_name'].str.contains(str(movie), regex=True)].shape[0] > 1):
        movie_data = {'collection': movie,
                      'count_movies': data.loc[data['collection_name'].str.contains(str(movie), regex=True), 'id'].count(),
                      'total_revenue': data.loc[data['collection_name'].str.contains(str(movie), regex=True), 'revenue'].sum(),
                      'mean_revenue': data.loc[data['collection_name'].str.contains(str(movie), regex=True), 'revenue'].mean()}
        return movie_data     
    else: return ('The movie you selected its not part of a collection.')  
print(collection('Toy Story'))
print(collection('jumanji'))
```

### def peliculas_pais( Pais: str ) ###
: Se ingresa un país (como están escritos en el dataset, no hay que traducirlos!), retornando la cantidad de peliculas producidas en el mismo.
                    Ejemplo de retorno: Se produjeron X películas en el país X

```python
def movies_country(country):
    movie_data = {f'total of movies produced in {country}': data.loc[data['prod_country'] == str(country), 'id'].count()}
    return movie_data
movies_country('United States of America')
```

### def productoras_exitosas( Productora: str ): ###
 Se ingresa la productora, entregandote el revunue total y la cantidad de peliculas que realizo.
                    Ejemplo de retorno: La productora X ha tenido un revenue de x

```python
def prod_companies_success(company):
    data.loc[data['prod_companies'].isna(), 'prod_companies'] = 'None'
    movie_data = {'company': company,
                  'total_movies': data.loc[data['prod_companies'].str.contains(str(company), regex=True), 'id'].count(),
                  'total_revenue': data.loc[data['prod_companies'].str.contains(str(company), regex=True), 'revenue'].sum()}
    return movie_data
prod_companies_success('Warner Bros.')               
```

### def get_director( nombre_director ): ### 
Se ingresa el nombre de un director que se encuentre dentro de un dataset debiendo devolver el éxito del mismo medido a través del retorno. Además, deberá devolver el nombre de cada película con la fecha de lanzamiento, retorno individual, costo y ganancia de la misma, en formato lista.

```python
def director_data(director):
    movies_data_list = []
    movies_data_list = []
    for i in range(data.loc[data['director'] == str(director), 'title'].shape[0]):
        movies_data_list.append([])
        movies_data_list[i] = {'title': data.loc[data['director'] == str(director), 'title'].values[i],
                                 'release_date': data.loc[data['director'] == str(director), 'release_date'].values[i],
                                 'budget': data.loc[data['director'] == str(director), 'budget'].values[i],
                                 'revenue': data.loc[data['director'] == str(director), 'revenue'].values[i],
                                 'return': data.loc[data['director'] == str(director), 'return'].values[i]}
        director_data = {'director':str(director),
                         'total_movies_return': data.loc[data['director'] == str(director), 'return'].sum(),
                         'movies_data': movies_data_list}
    return(director_data)
director_data('Joe Johnston')
```
